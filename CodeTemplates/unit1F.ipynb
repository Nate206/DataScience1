```python
df['rating'] = df.groupby('author')['rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
print(df.isna().sum()),
```

```python
average_rating = df['rating'].mean()
df['rating'].fillna(average_rating, inplace=True)
```

```python
print(df.isna().sum()),
```

```python
mean_price = df['rating'].mean()
print( mean_price)
```

```python
median_price = df['book price'].median()
print( median_price)
```

```python
mode_price = df['genre'].mode()
if len(mode_price) == 0:
    print("No mode found.")
else:
    print("Mode Book Price:", mode_price.values)
```

```python
import matplotlib.pyplot as plt

genre_counts = df['genre'].value_counts()
genre_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Book Genres')
plt.xlabel('Genres')
plt.ylabel('Count')
plt.show()
```

```python
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Genre Proportions')
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x='genre', y='rating', data=df, palette='viridis')
plt.title('Box Plot of Ratings by Genre')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='author', hue='genre', data=df, palette='Set2')
plt.title('Count of Books by Genre for Each Author')
plt.xlabel('Authors')
plt.ylabel('Count')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
sns.violinplot(x='genre', y='rating', data=df, palette='muted')
plt.title('Violin Plot of Ratings by Genre')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'genre' is a categorical variable, and 'rating' and 'book_price' are numerical variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rating', y='book price', hue='genre', data=df, palette='Set2', s=100)
plt.title('Scatter Plot of Ratings vs. Book Prices by Genre')
plt.xlabel('Rating')
plt.ylabel('Book Price')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
