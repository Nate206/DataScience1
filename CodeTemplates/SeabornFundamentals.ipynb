```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("compas.csv")
```

```python
sns.countplot(x=df["sex"])
```

```python
sns.countplot(x=df["age_cat"])
```

```python
sns.histplot(df['age'], bins=8, kde=True, color='skyblue')
```

```python
df['total_time_jail']=(df["c_jail_out"] - df["c_jail_in"]).dt.days
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=df, x= "total_time_jail", y = "age")

plt.xlabel("Days in Jail")
plt.title('Days in Jail After First Charge')
```
